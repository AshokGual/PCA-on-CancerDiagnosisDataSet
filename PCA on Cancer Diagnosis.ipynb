```python
# diagnosis consist of categorical value so transforming it into numerical values

def generatenum():
    return cancer['diagnosis'].map({'M':1,'B':0})
cancer['diagnosis']=generatenum()
```

```python
cancer.head()
```

```python
cancer['diagnosis'].value_counts()
```

```python
#Extract the independent variable from dataset

X=cancer.loc[:,cancer.columns!='diagnosis']
X.head()
```

```python
#Extract the dependent variable from dataset
Y=cancer.loc[:,cancer.columns=='diagnosis']
Y.head()
```

```python
# scalling of dataset X into a similar range
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X)
x=scaler.transform(X)
x=pd.DataFrame(x)
x.head()
```

```python
#split the X and Y into train and test dataset
from sklearn.model_selection import train_test_split

def split():
    return train_test_split(X,Y,test_size=0.25, random_state=1)
X_Train,X_Test,Y_Train,Y_Test=split()
```

```python
print(X_Train.shape)
print(X_Test.shape)
print(Y_Train.shape)
print(Y_Test.shape)
```

Implement the RandomForestClassifier Model

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(random_state=1)
rfc.fit(X_Train,Y_Train)
y_predict=rfc.predict(X_Test)
print(y_predict)
print(Y_Test)
```

Find the Accuracy of the model

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(Y_Test,y_predict)
print('Accuracy of the RandomForestClassifier Model is :',accuracy)
```

Apply confusion matrix to determine the performance of RFC model

```python
from sklearn.metrics import confusion_matrix
confusion_accuracy=confusion_matrix(Y_Test,y_predict)
print(confusion_accuracy)
plt.matshow(confusion_accuracy)
plt.colorbar()
plt.show()
```

Here we got the accuracy of 95.10% by applying RFC model on 31 Features

# LETS apply PCA Model on this data set and then we implement the RFC model and check the accuracy

```python
#Apply PCA and calculate the % of variance
from sklearn.decomposition import PCA
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

def pcadecom():
    pca=PCA(n_components=31,random_state=1)
    pca.fit(x)
    xi=np.arange(0,31,1)
    y=np.cumsum(pca.explained_variance_ratio_)
    plt.figure(figsize=(12,6))
    plt.ylim(0.3,1.1)
    #var=np.cumsum(np.round(pca.explained_variance_ratio_,decimals=4)*100)
   
    plt.xticks(np.arange(0,32))
    plt.plot(xi,y,marker='o',linestyle='--')
        
    return plt.show()

pcadecom()
```
