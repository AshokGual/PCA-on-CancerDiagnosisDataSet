```python
# diagnosis consist of categorical value so transforming it into numerical values

def generatenum():
    return cancer['diagnosis'].map({'M':1,'B':0})
cancer['diagnosis']=generatenum()
```

```python
#Extract the independent variable from dataset

X=cancer.loc[:,cancer.columns!='diagnosis']
X.head()
```

```python
#Extract the dependent variable from dataset
Y=cancer.loc[:,cancer.columns=='diagnosis']
Y.head()
```

```python
#split the X and Y into train and test dataset
from sklearn.model_selection import train_test_split

def split():
    return train_test_split(X,Y,test_size=0.25, random_state=1)
X_Train,X_Test,Y_Train,Y_Test=split()
```

```python
print(X_Train.shape)
print(X_Test.shape)
print(Y_Train.shape)
print(Y_Test.shape)
```

```python
# scalling of dataset X into a similar range
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X)
x=scaler.transform(X)
```

```python
x.shape
```

```python
#Apply PCA and calculate the % of variance
from sklearn.decomposition import PCA
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

def pcadecom():
    pca=PCA(n_components=32,random_state=1)
    pca.fit(x)
    xi=np.arange(0,32,1)
    y=np.cumsum(pca.explained_variance_ratio_)
    plt.figure(figsize=(12,6))
    plt.ylim(0.3,1.1)
    #var=np.cumsum(np.round(pca.explained_variance_ratio_,decimals=4)*100)
   
    plt.xticks(np.arange(0,32))
    plt.plot(xi,y,marker='o',linestyle='--')
        
    return plt.show()

pcadecom()
    
    

```

```python
from sklearn.decomposition import PCA

def reduced():
    pca1=PCA(n_components=0.95,random_state=1)
    pca1.fit(x)
    return pca1.transform(x)
x1=reduced()
x1.shape
```

```python
#split x and y in to train and test

from sklearn.model_selection import train_test_split

def split1():
    return train_test_split(x1,Y, test_size=0.25, random_state=1)
x1_train,x1_test,Y_train,Y_test=split1()    
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(x1_train,Y_train)
y_predict=rfc.predict(x1_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,y_predict))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test,y_predict))
```
